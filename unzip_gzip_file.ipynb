```python
import os
import argparse
import pandas as pd
import polars as pl

# Define constants
AWS_ACCESS_KEY_ID = os.getenv('AWS_ACCESS_KEY_ID')
AWS_SECRET_ACCESS_KEY = os.getenv('AWS_SECRET_ACCESS_KEY')
AWS_REGION = os.getenv('AWS_REGION')
BUCKET_NAME = 'your_bucket_name_here'
TABLE_NAME = 'your_table_name_here'

# Define functions
def extract_data(new_products_only=False):
    """
    Extract data from the data warehouse.
    If new_products_only=True, only extract data for new products (after first run).
    """
    if new_products_only:
        # Read the table that was created in the previous run
        try:
            df = pd.read_parquet(f's3://{BUCKET_NAME}/{TABLE_NAME}.parquet.gzip')
        except FileNotFoundError:
            raise ValueError("Couldn't find previous table to extract new products from.")
        # Get a list of existing product IDs
        existing_product_ids = df['product_id'].unique()
        # Extract only rows with new product IDs
        query = f'product_id.not_in([{", ".join(existing_product_ids)}])'
        df = pl.read_sql_query(f"SELECT * FROM my_table WHERE {query}")
    else:
        # Extract all data
        df = pl.read_sql_table('my_table')
    return df

def process_data(df):
    """
    Process the data as necessary.
    """
    # Do some processing
    df = df.groupby('category').agg({
        'price': 'mean',
        'quantity': 'sum'
    })
    return df

def deposit_data(df):
    """
    Deposit the processed data into an S3 bucket.
    """
    # Write the dataframe to a parquet file
    df.to_parquet(f's3://{BUCKET_NAME}/{TABLE_NAME}.parquet.gzip', compression='gzip')

# Define command-line arguments
parser = argparse.ArgumentParser()
parser.add_argument('--new-products', action='store_true',
                    help='Extract data for new products only.')
args = parser.parse_args()

if __name__ == '__main__':
    # Set up AWS credentials
    pl.aws.connect_access_key(
        AWS_ACCESS_KEY_ID,
        AWS_SECRET_ACCESS_KEY,
        region=AWS_REGION
    )

    # Extract data
    df = extract_data(args.new_products)

    # Process data
    df = process_data(df)

    # Deposit data
    deposit_data(df)
```
